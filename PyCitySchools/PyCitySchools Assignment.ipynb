```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_csv('schools_complete.csv').sort_values(["school_name"], ascending=True)
df2=pd.read_csv('students_complete.csv').sort_values(["school_name"], ascending=True)
```

```python
df1.head()
```

```python
df2.head()
```

# District Summary


```python
Total_School=df1['School ID'].count()
```

```python
Total_Student=df2['Student ID'].nunique()
```

```python
Total_Budget=df1.budget.sum()
```

```python
Average_Math_Score=df2['math_score'].mean()
```

```python
Average_Reading_Score=df2.reading_score.mean()
```

```python
Pass_Math = (df2["math_score"] >= 70).value_counts()[True]
Pass_Math_Perc=round((Pass_Math/Total_Student)*100, 2)
```

```python
Pass_Reading=(df2["reading_score"] >= 70).value_counts()[True]
Pass_Reading_Perc=round((Pass_Reading/Total_Student)*100, 2)
```

```python
Overall_Passing_Perc=round((Pass_Math_Perc + Pass_Reading_Perc)/2, 2)
```

```python
District_Summary=pd.DataFrame({"Total Schools":[Total_School], "Total Students":[Total_Student], "Total Budget":[Total_Budget],
                               "Average Math Score":[Average_Math_Score], "Average Reading Score":[Average_Reading_Score], 
                              "% Passing Math":[Pass_Math_Perc], "% Passing Reading":[Pass_Reading_Perc],
                               " Overall Passing % ":[Overall_Passing_Perc]})
District_Summary 
```

# School Summary

```python
dfmerge = pd.merge(df2, df1, how="left", on=["school_name", "school_name"])
dfmerge.head()
```

```python
groupedschool=dfmerge .groupby('school_name')
groupedschool
```

```python
SchoolType=groupedschool["type"].first()
TotalStudents=groupedschool['Student ID'].count()
TotalSchoolBudget=groupedschool['budget'].sum()
totalSchoolBudget = groupedschool["budget"].first()
PerStudentBudget=totalSchoolBudget/TotalStudents
AverageMathScore=groupedschool['math_score'].mean()
AverageReadingScore=groupedschool['reading_score'].mean()
```

```python
passingMath = dfmerge[dfmerge ["math_score"] >= 70].groupby(["school_name"])["math_score"].count()
percentagePassingMath = passingMath / TotalStudents * 100
passingRead = dfmerge[dfmerge ["reading_score"] >= 70].groupby(["school_name"])["reading_score"].count()
percentagePassingReading = passingRead / TotalStudents * 100
percentageOverallPassingRate = (percentagePassingMath + percentagePassingReading) / 2
```

```python
School_Summary=pd.DataFrame({"School Type":SchoolType, "Total Students":TotalStudents, "Total School Budget":TotalSchoolBudget, 
                            "Per Student Budget":PerStudentBudget, "Average Math Score":AverageMathScore, 
                             "Average Reading Score":AverageReadingScore, "% Passing Math": percentagePassingMath,
                            "% Passing Reading":percentagePassingReading, "% Overall Passing": percentageOverallPassingRate})
```

```python
School_Summary.head(5)
```

# Top Performing Schools (By % Overall Passing)


```python
TopPerforming=School_Summary.sort_values(["% Overall Passing"], ascending=False)
TopPerforming.head(5)
```

# Bottom Performing Schools (By % Overall Passing)

```python
BottomPerforming=School_Summary.sort_values(["% Overall Passing"], ascending=True)
BottomPerforming.head(5)
```
